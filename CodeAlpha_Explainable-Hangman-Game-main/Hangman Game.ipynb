```python
import random

print("HANGMAN GAME | Developed by Sarvesh")

stages = [
    """
     |
     |
     |
     |
     |
    ===
    """,
    """
     |
     |
     O
     |
     |
    ===
    """,
    """
     |
     |
     O
     |
     |
    ===
    """,
    """
     |
     |
     O
    /|
     |
    ===
    """,
    """
     |
     |
     O
    /|\\
     |
    ===
    """,
    """
     |
     |
     O
    /|\\
    /
    ===
    """,
    """
     |
     |
     O
    /|\\
    / \\
    ===
    """
]

words = ["python", "java", "hangman", "coding", "program"]
random.shuffle(words)
word = words[0]

guessed = []
wrong_guessed = []
attempts = 6
total_guesses = 0
vowel_guesses = 0
consonant_guesses = 0

vowels = "aeiou"

print("WELCOME TO THE EXPLAINABLE HANGMAN WORD GUESSING GAME")
print("YOU HAVE 6 INCORRECT ATTEMPTS")

while attempts > 0:
    print(stages[6 - attempts])

    display = ""
    for ch in word:
        if ch in guessed:
            display += ch + " "
        else:
            display += "_ "

    print("Word:", display.strip())
    print("Wrong Letters:", wrong_guessed)
    print("Remaining Attempts:", attempts)

    if "_" not in display:
        print("\nCONGRATULATIONS! YOU WON THE GAME.")
        break

    guess = input("Enter a letter: ").lower()

    if len(guess) != 1 or not guess.isalpha():
        print("INVALID INPUT. ENTER A SINGLE LETTER.")
        continue

    if guess in guessed or guess in wrong_guessed:
        print("LETTER ALREADY GUESSED. TRY A NEW LETTER.")
        continue

    total_guesses += 1

    if guess in vowels:
        vowel_guesses += 1
    else:
        consonant_guesses += 1

    if guess in word:
        guessed.append(guess)
        print("CORRECT GUESS.")
    else:
        wrong_guessed.append(guess)
        attempts -= 1
        print("WRONG GUESSED LETTER. TRY A NEW LETTER.")

print("\nGAME ANALYSIS REPORT")
print("--------------------")
print("Total Guesses:", total_guesses)
print("Correct Guesses:", len(guessed))
print("Wrong Guesses:", len(wrong_guessed))

if vowel_guesses > consonant_guesses:
    style = "VOWEL-FOCUSED GUESSING"
elif consonant_guesses > vowel_guesses:
    style = "CONSONANT-FOCUSED GUESSING"
else:
    style = "BALANCED GUESSING"

print("Guessing Style:", style)

if attempts == 0:
    print(stages[6])
    print("\nGAME OVER!")
    print("CORRECT WORD:", word)
```
