```python
import time
import random

def sarvmind_x():
    print("ü§ñ SarvMind X: Hello Sarvesh.")
    print("ü§ñ SarvMind X: I don't just reply ‚Äî I explain, reflect, and adapt.")
    print("ü§ñ SarvMind X: Type 'bye' to exit.\n")

    last_topic = None
    last_time = time.time()

    responses = {
        "greeting": [
            "Hello. I'm here with you.",
            "Hi. You have my attention.",
            "Hey. Let's talk."
        ],
        "emotional": [
            "That sounds emotionally heavy.",
            "It seems this is affecting you.",
            "I sense something important behind your words."
        ],
        "neutral": [
            "I see. Tell me more.",
            "Understood. Go on.",
            "I'm listening."
        ]
    }

    while True:
        user_input = input("You: ").strip()
        current_time = time.time()
        pause_duration = current_time - last_time
        last_time = current_time

        # Exit
        if user_input.lower() in ["bye", "exit", "quit"]:
            print("ü§ñ SarvMind X: Goodbye. Thank you for the conversation.")
            break

        # üîï Silence / Pause Awareness
        if pause_duration > 6:
            print("ü§ñ SarvMind X: I noticed you took a pause.")
            print("üß† Reason: Extended silence often means hesitation or reflection.")

        # Greeting detection
        if any(word in user_input.lower() for word in ["hi", "hello", "hey"]):
            response = random.choice(responses["greeting"])
            reason = "Detected greeting keywords like 'hi/hello'."
            topic = "greeting"

        # Emotional signal detection
        elif any(word in user_input.lower() for word in
                 ["tired", "sad", "stress", "hard", "alone", "confused"]):
            response = random.choice(responses["emotional"])
            reason = "Detected emotionally weighted words."
            topic = "emotion"

        # Fallback (open conversation)
        else:
            response = random.choice(responses["neutral"])
            reason = "No strong intent detected; keeping conversation open."
            topic = "general"

        # üó£ Main Response
        print(f"ü§ñ SarvMind X: {response}")

        # üß† Explainability Layer
        print(f"üß† Reason: {reason}")

        # üîç Self-Critic / Reflection Layer
        if topic == "emotion":
            print("üîé Self-Check: Was my response empathetic enough?")
            improved = response + " If you want to talk about it, I'm here."
            print(f"üîß Improved Response: {improved}")

        elif topic == "general" and last_topic == "emotion":
            print("üîé Self-Check: User previously expressed emotion ‚Äî maintaining sensitivity.")

        last_topic = topic


# Run SarvMind X
sarvmind_x()
```
